# Student Depression & Lifestyle Risk Analysis

## Module 2: Lifestyle and Well-being Risk Detection

This notebook trains baseline classifiers for the Student Depression dataset and outputs `lifestyle_score` as a well-being risk signal. It is a research evaluation pipeline, not a medical diagnosis workflow.

```python
import pandas as pd

from src.preprocessing.preprocess_depression import preprocess_depression
from src.models.lifestyle_model import LifestyleRiskModel
```

```python
RAW_DATA_PATH = 'data/raw/student_depression/student_depression.csv'
PROCESSED_DATA_PATH = 'data/processed/module2_lifestyle.csv'
METRICS_PATH = 'results/metrics_module2.csv'

X, y = preprocess_depression(RAW_DATA_PATH)
pd.concat([X, y], axis=1).to_csv(PROCESSED_DATA_PATH, index=False)
print(f'Rows: {len(X):,}')
print(f'Positive rate: {y.mean():.3f}')
print(f"Processed file saved to: {PROCESSED_DATA_PATH}")
X.head()
```

```python
model = LifestyleRiskModel(random_state=42)
comparison_df, holdout = model.compare_models(X, y)
best_result = comparison_df.iloc[0]
metrics_df = pd.DataFrame(
    {
        'metric': ['model_name', 'accuracy', 'precision', 'recall', 'f1_score', 'roc_auc'],
        'value': [
            best_result['model_name'],
            best_result['accuracy'],
            best_result['precision'],
            best_result['recall'],
            best_result['f1_score'],
            best_result['roc_auc'],
        ],
    }
)
metrics_df.to_csv(METRICS_PATH, index=False)
print(f"Metrics saved to: {METRICS_PATH}")
comparison_df
```

```python
holdout_predictions = holdout['X_test'].copy()
holdout_predictions['true_label'] = holdout['y_test'].values
holdout_predictions['predicted_label'] = model.predict(holdout['X_test'])
holdout_predictions['lifestyle_score'] = model.predict_scores(holdout['X_test']).values
holdout_predictions.head(10)
```

### Ethical note

Sensitive variables in this dataset should only be used for research evaluation. In real deployment, the system should rely on consent-based and necessary data only.
